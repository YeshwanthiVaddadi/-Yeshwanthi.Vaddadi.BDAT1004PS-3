```python
#Question 1:

"""Occupations
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address.
Step 3. Assign it to a variable called users
Step 4. Discover what is the mean age per occupation"""

import pandas as pd
import numpy as np

users= pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|', index_col='user_id')
users.groupby('occupation').age.mean()
```

```python
#Question 1: Step 5. Discover the Male ratio per occupation and sort it from the most to the least


ratio = users.pivot_table(index='occupation', columns='gender', aggfunc='size', fill_value = 0)

Total = ratio[['F', 'M']].sum(axis = 1)
ratio['MaleRatio'] = round(100 * ratio['M'] / Total, 1)
ratio['FemaleRatio'] = round(100 * ratio['F'] / Total, 1)

ratio.sort_values(by='MaleRatio', ascending=False)
```

```python
#Question 1: Step 6. For each occupation, calculate the minimum and maximum ages

users.groupby('occupation').age.agg(['min','max'])
```

```python
#Question 1 Step 7. For each combination of occupation and sex, calculate the mean age

users.groupby(['occupation', 'gender']).age.mean()
```

```python
#Question 9

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load data from file
data = pd.read_csv(r"E:\BDAT_Georgian\BDAT_1004_Data Programming\Assignments\roman-emperor-reigns.csv")

# Count number of people assassinated and calculate percentage
assassinated_count = (data['Cause_of_Death'].str.contains('assassinated', case=False)).sum()
total_count = len(data)
assassinated_percent = round((assassinated_count / total_count) * 100, 2)

# Calculate percentage of people who died of natural causes or other reasons
other_percent = round((total_count - assassinated_count) / total_count * 100, 2)

# Data
labels = ['Assassinated', 'Other']
sizes = [assassinated_percent, other_percent]
colors = ['#FF5733', '#007ACC']
explode = (0.1, 0)

# Create a larger pie chart with different colors
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
ax.axis('equal')
plt.title('Fraction of Roman data Assassinated')

# Display the chart
plt.show()
```
